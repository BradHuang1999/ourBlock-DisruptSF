```python
import pandas as pd
```

```python
df = pd.read_json('sample-reports.json')
```

```python
df
```

```python
import random as rd
# -0.001 ~ +0.001
```

```python
def locaShift(v):
    v['lat'] += round(rd.uniform(-0.001,0.001), 10)
    v['lon'] += round(rd.uniform(-0.001,0.001), 10)
    return v

df = df.apply(locaShift, axis=1)
```

```python
df.to_json('Cleaned-Data.json', orient='records')
```
